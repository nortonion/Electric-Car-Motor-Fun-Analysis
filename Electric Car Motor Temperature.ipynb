###### We realize that the current and voltage (i_d, i_q, u_d, u_q) drive the motor_speed and torque from the dataset details. Therefore, let's see if we can make a model to predict motor speed and torque of the motor using current and voltage values in d,q dimensions.

```python
#independent variables are the current and voltage values in d,q dimensions
X = df.iloc[:,[2,3,6,7]].values
```

```python
#dependent variable we look at here is motor speed
y = df.iloc[:,4].values
```

```python
"""
Import the traing_test_split module (we import it here so that we understand how we can freely split dataset
into training and testing data) 
Then we split the dataset into training and test sets
"""
from sklearn.model_selection import train_test_split
#We split training:test in ratio of 80%:20%; also random_state = 0 makes it easy for someone else to replicate by
#using specific random seed
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
df.head(10)
```

```python
# Feature Scaling
"""
We use feature scaling so that we do not have variables given more weight because of their magnitude.
We standardize the data so that the mean is 0 and the standard deviation of the new distribution is 1.
"""
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1,1))
```

##### Let's first try multiple linear regression!

```python
#Fitting Multiple Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Use linear regression model to predict
y_pred = regressor.predict(X_test)
```

```python
#R^2 of model... wow, 90% is very high!
regressor.score(X_test,y_test)
```

### 90.6% is a very high score! Let's use k-fold cross validation to check.

```python
#Applying k-fold Cross Validation
from sklearn.model_selection import cross_val_score
#We use 10-fold cross validation so cv = 10
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10)
```

```python
#We get a table of scores because we test different chunks
accuracies
```

```python
accuracies.mean()
```

```python
#standard deviation of the accuracies
accuracies.std()
```

### Let's try clustering motor speed and the torque and visualize it!

```python
#We will cluster the motor speed and the torque
X = df.iloc[:,[4,5]].values
```

```python
"""
We will find optimal amount of clusters using WCSS (within cluster sums of squares).
Then we will plot it and see where the "elbow" is.
"""
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 10, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('# of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Again, we see that the number of observations takes overwhelmingly long to run! This is for k-clustering!
